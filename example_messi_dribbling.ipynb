# 3D Depth Mapping via MiDaS

### Monocular Depth Estimation with a Single Image (MiDaS)

```python
import os

from depth_estimation import DepthMapper
from image_processing import ImageProcessor
```

## MiDaS x PyTorch to create depth map 

_Example: Lionel Messi dribbling vs multiple defenders_

```python
MAPPING_LVL = 2
PHOTO = "messi_vs_everyone.jpg"

photo_path = os.path.join(os.getcwd(), "images", PHOTO)
```

<img src="photos/messi_vs_everyone.jpg"   />


### Execute mapping algorithm

```python
depthmapper = DepthMapper(MAPPING_LVL)
processed_image = ImageProcessor.from_file(photo_path)
unprocessed_depthmap = depthmapper.map(processed_image.image)
processed_depthmap = ImageProcessor.from_array(unprocessed_depthmap)
```

### Display depth map (image)

```python
processed_depthmap.display()
```
